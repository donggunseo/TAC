```python
import json

def tsv_to_json(tsv_path, json_path):
    data_list = []

    # TSV 파일 읽기
    with open(tsv_path, 'r', encoding='utf-8') as f:
        for line in f:
            parts = line.rstrip("\n").split("\t")
            if len(parts) != 2:
                continue  # 혹은 에러 처리
            
            input_text, output_text = parts
            data_list.append({"input": input_text, "output": output_text})

    # JSON 파일로 저장
    with open(json_path, 'w', encoding='utf-8') as f:
        json.dump(data_list, f, ensure_ascii=False, indent=2)

    print(f"Saved {len(data_list)} items to {json_path}")

```

```python
task = "TFU"
dir_path = f"./dataset/StylePTB_{task}/"
for split in ["train", "valid", "test"]:
    tsv_to_json(f"{dir_path}{split}.tsv", f"{dir_path}{split}.json")
```
