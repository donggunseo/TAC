```python
from transformers import AutoTokenizer
import json

tokenizer = AutoTokenizer.from_pretrained('meta-llama/Llama-3.1-8B')
```

```python
task = "clinc150"
```

```python
with open(f"./dataset/{task}/label_list.json", "r") as f:
    label_list = json.load(f)
```

```python
from collections import defaultdict

d = defaultdict(int)
for l in label_list:
    t = tokenizer.tokenize(l)
    d[len(t)]+=1
```

```python
with open(f"./result_extended/{task}/llama3.1-8b/extended_act_converter_result.json", "r") as f:
    res = json.load(f)
```

```python
def last_matching_index(list1, list2):
    limit = min(len(list1), len(list2))
    last_match = -1

    for i in range(limit):
        if list1[i] == list2[i]:
            last_match = i
        else:
            break

    return last_match
```

```python
oof_cnt = 0
d2 = defaultdict(int)
for item in res['res']:
    if item['pred'] not in label_list:
        oof_cnt+=1
        print(f"pred : {item['pred']} | gt : {item['gt']}")
        print(last_matching_index(tokenizer.tokenize(" "+item['pred']), tokenizer.tokenize(" "+item['gt'])))
        print("__________")
```

```python
print(oof_cnt/len(res['res']))
```
